```
import pandas as pd
from sklearn.datasets import load_iris
iris= load_iris()
df=pd.DataFrame(
    iris['data'],
    columns=iris["feature_names"]
)
df["ans"]=iris["target"]
df.to_csv("iris.csv",encoding="utf-8",index=False)
df
```

```
import numpy as np
from sklearn.model_selection import train_test_split
I_=df.drop(["ans"],axis=1)
O_=df["ans"]
I_,O_=np.array(I_),np.array(O_)
train_I,test_I,train_O,test_O=train_test_split(I_,O_,test_size=0.1)
```

```
from sklearn.tree import DecisionTreeClassifier
len_=len(df)
clf=DecisionTreeClassifier(min_samples_leaf=3) # max_depth=2
clf.fit(train_I,train_O)
```

```
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
plt.figure(figsize=(20,20))
tree_=plot_tree(
    clf,
    feature_names=iris["feature_names"],
    class_names=iris["target_names"],
    filled=True
)
plt.show()
```

```
from sklearn.metrics import accuracy_score
pre=clf.predict(test_I)
accuracy_score(pre,test_O)
```
